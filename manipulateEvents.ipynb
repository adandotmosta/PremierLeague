```python
#get player red cards and yellow cards
# Concatenate all DataFrames in 'dfs' to create a single DataFrame
combined_df = pd.concat(dfs.values(), ignore_index=True)

# Filter the combined DataFrame for "Shot" events only
cards = combined_df[(combined_df['Event Name'] == 'Yellow Card') | (combined_df['Event Name'] == 'Red Card')]

# Select only the relevant columns
cards = cards[[ 'Player1 Name', 'Player1 Team',"Match Name","Event Name"]]

# Reset the index of the DataFrame
cards = cards.reset_index(drop=True)
```

```python
# event name contain red card or yellow card, i want to sum yellow cards in column yellow card and red cards in red card
cards["Yellow Card"] = 0
cards["Red Card"] = 0
for index, row in cards.iterrows():
    if row["Event Name"] == "Yellow Card":
        cards.loc[index, "Yellow Card"] = 1
    else:
        cards.loc[index, "Red Card"] = 1 
```

```python
# now sum yellow cards and red cards
cards = cards.groupby(["Player1 Name", "Player1 Team"]).agg({"Yellow Card": "sum", "Red Card": "sum"}).reset_index()
```
