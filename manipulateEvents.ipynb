```python
players_coords = {}
for df in dfs.values():
    for index, row in df.iterrows():
        player = row["Player1 Name"]
        event = row["Event Name"]
        x = row["X"]  # Assurez-vous que vous avez les colonnes X et Y dans votre dataframe
        y = row["Y"]
        
        # Vérification si le joueur existe déjà dans le dictionnaire
        if player not in players_coords:
            players_coords[player] = {}
        
        # Vérification si l'événement existe déjà pour ce joueur
        if event not in players_coords[player]:
            players_coords[player]["Shot"] = []
        
        # Ajouter les coordonnées (x, y) pour cet événement spécifique du joueur
        players_coords[player]["Shot"].append((x, y))
```

```python
# Concatenate all DataFrames in 'dfs' to create a single DataFrame
combined_df = pd.concat(dfs.values(), ignore_index=True)

# Filter the combined DataFrame for "Shot" events only
shot_df = combined_df[combined_df['Event Name'] == 'Shot']

# Select only the relevant columns
shot_df = shot_df[['X', 'Y', 'Half', 'Player1 Name', 'Player1 Team',"Match Name"]]

# Reset the index of the DataFrame
shot_df = shot_df.reset_index(drop=True)

# Display the DataFrame
print(shot_df)
```
