```python
import re
def extract_home_away_teams(match_str: str):
    """
    Extracts the home and away teams from a match string.
    Args:
    - match_str (str): Match string like '03.03 Wigan Athletic v Swansea City - Events.csv'
    
    Returns:
    - (home_team, away_team) tuple
    """
    # Remove the date at the beginning of the string using regex
    match_str_cleaned = re.sub(r'^\d{2}\.\d{2}\.\d{2,4} ', '', match_str.split(" -")[0])  # Removing the date part
    
    # Split the cleaned string based on ' v '
    match_split = match_str_cleaned.split(" v ")
    
    if len(match_split) == 2:
        home_team = match_split[0].strip()  # Home team
        away_team = match_split[1].strip()  # Away team
        return home_team, away_team

players = {}
wins = 0
a_domicile = 0
team = "Manchester City"  # Team to analyze
total_xg = 0  # Variable to accumulate total xG
for df in dfs.values():
    match = df
    match_name = df['Match Name'].iloc[0]
    print(f"Processing match: {match_name}")
    
    home_team, away_team = extract_home_away_teams(match_name)
    print(f"Home team: {home_team}, Away team: {away_team}")

    # Increment a_domicile if the team is the home team
    

    
    # Count wins

    if match.iloc[0]['Result'] == 'W' and team in [home_team, away_team]:
        wins += 1

        if home_team == team:
            a_domicile += 1
print(a_domicile)
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Initialisation des données des joueurs
players = {}
selected_attributes = [
    "Goals", "xGoals Shot", "DP Passes Made", "Cr Crosses Made",
    "CA Regains", "HP Regains", "BMP+", "Possessions", "On-pitch xG Generated",
    "ST Involvement", "BU Involvement", "Ma Involvement", "Player Points",
    "Average X", "Average Y"
]

sum_attributes = ["Goals", "xGoals Shot", "DP Passes Made", "Cr Crosses Made", 
                  "CA Regains", "HP Regains", "BMP+", "On-pitch xG Generated"]

mean_attributes = ["Possessions", "ST Involvement", "BU Involvement", "Ma Involvement", 
                   "Player Points", "Average X", "Average Y"]

# Comptabilisation des matchs joués et stockage des équipes
player_matches = {}
player_teams = {}

# Agrégation des données
for df in dfs.values():
    for _, row in df.iterrows():
        player_name = row["Player Name"]
        
        if player_name not in players:
            players[player_name] = {attr: 0 for attr in selected_attributes}
            player_matches[player_name] = 0
            player_teams[player_name] = row["Team"]  # Stocker l'équipe du joueur

        player_matches[player_name] += 1  # Compte le nombre de matchs

        for attr in sum_attributes:
            if attr in row:
                players[player_name][attr] += row[attr]  # On somme ces attributs

        for attr in mean_attributes:
            if attr in row:
                players[player_name][attr] += row[attr]  # On accumule pour la moyenne

# Moyenne pour les stats choisies
for player, stats in players.items():
    for attr in mean_attributes:
        stats[attr] /= player_matches[player]  # Moyenne par match

# Conversion en DataFrame
players_df = pd.DataFrame.from_dict(players, orient="index")

# Ajouter "Player Name" comme colonne au lieu de l'index
players_df.reset_index(inplace=True)
players_df.rename(columns={"index": "Player Name"}, inplace=True)

# Ajouter la colonne "Team"
players_df["Team"] = players_df["Player Name"].map(player_teams)

# **Suppression des gardiens (Average X < -39)**
# Correction de la condition de filtrage des gardiens
players_df = players_df[(players_df["Average X"] >= -39) & (players_df["Player Name"] != "Paul ROBINSON")]


# Sélectionner uniquement les attributs à normaliser (exclure Average X et Average Y)
attributes_to_normalize = [attr for attr in selected_attributes if attr not in ["Average X", "Average Y"]]

# **Renormalisation après suppression des gardiens**
scaler = MinMaxScaler()
players_df[attributes_to_normalize] = scaler.fit_transform(players_df[attributes_to_normalize])

# Afficher les premières lignes
print(players_df.head())
```

```python
# convert into csv
players_df.to_csv('players_performance.csv')
```

```python
def evolutionary_stat(self, player_name, team_name):
    """
    Returns the evolution of the player's stats over time.
    
    Args:
    - player_name (str): Name of the player
    - team_name (str): Name of the team
    
    Returns:
    - DataFrame containing the player's stats over time
    """
    # Load match files
    files = self.load_match_files()
    
    # Initialize an empty list to accumulate the player's stats over time
    player_stats = []
    
    # Iterate through each match to accumulate the player's stats
    for match in files:
        print(match)
        if team_name in match:
            print("team_name",team_name)
            # Load player data for the current match
            data = self.load_player_data(match + "- Players.csv")
            
            # Filter data to get the player's stats
            player_data = data[data['Player Name'] == player_name]
            
            
            if not player_data.empty:
                goals = player_data['Goals'].values[0]
                xg = player_data['xGoals Shot'].values[0]
                minutes = player_data['Minutes Played'].values[0]
                
                # Append the player's stats for the match to the list
                player_stats.append({'Match': match, 'Goals': goals, 'xG': xg, 'Minutes Played': minutes})
    
    # Convert the list of stats into a DataFrame
    player_stats_df = pd.DataFrame(player_stats)
    
    return player_stats_df

def evolution_by_month_match(self, player_name, team_name, aggregation="month"):
    df = self.evolutionary_stat(player_name, team_name)

    if aggregation == "month":
        try:
            # Extraire la date (ex: "12.03.12" depuis "12.03.12 Arsenal v Newcastle United")
            df["Match"] = df["Match"].str.extract(r"(\d{2}\.\d{2}\.\d{2})")[0]

            # Convertir en datetime
            df["Match"] = pd.to_datetime(df["Match"], format="%d.%m.%y")

            # Extraire le mois sous forme de période (YYYY-MM)
            df["Month"] = df["Match"].dt.to_period("M")

            # Sélectionner uniquement les colonnes numériques pour l'agrégation
            numeric_cols = df.select_dtypes(include=["number"]).columns
            df = df.groupby("Month")[numeric_cols].sum().reset_index()

            # Convertir 'Month' en string pour éviter les erreurs JSON (Plotly, Streamlit)
            df["Month"] = df["Month"].astype(str)

            return df
        except Exception as e:
            print("Error during aggregation:", e)
            return df

    elif aggregation == "match":
        return df
```
